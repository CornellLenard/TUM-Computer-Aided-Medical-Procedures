## Part 1 - Spherical Displacements

In this part, we look at 3D datasets.

```python
## Load datasets from file
P_fix = np.loadtxt("data/faceX.txt")
P_mov = np.loadtxt("data/faceY.txt")
N = P_fix.shape[1]

## Okay, so this is what an input dataset in 3D could look like:
fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")
ax.scatter(P_fix[0,], P_fix[1,], P_fix[2,], c="r")
ax.scatter(P_mov[0,], P_mov[1,], P_mov[2,], c="b")
lim = 3
ax.set_xlim(2 - lim, 2 + lim)
ax.set_ylim(-lim, lim)
ax.set_zlim(-lim, lim)
plt.show()
```

```python
## Demean datasets..
mean_fix = np.mean(P_fix, axis=1, keepdims=True) #TODO compute mean
mean_mov = np.mean(P_mov, axis=1, keepdims=True) #TODO compute mean

P_fix_demean = P_fix - mean_fix #TODO demean
P_mov_demean = P_mov - mean_mov #TODO demean

# Plot
fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")
ax.scatter(P_fix_demean[0,], P_fix_demean[1,], P_fix_demean[2,], c="r")
ax.scatter(P_mov_demean[0,], P_mov_demean[1,], P_mov_demean[2,], c="b")
lim = 2
ax.set_xlim(-lim, lim)
ax.set_ylim(-lim, lim)
ax.set_zlim(-lim, lim)
plt.show()
```

Now, let's solve the spherical registration in 3D:

```python
# Assemble H matrix - we will use this multiple times
def compute_H(a, b):
    #TODO compute and return H
    return np.asmatrix(a) * np.asmatrix(b).T
    
## Method of Umeyama.
def spherical_umeyama(x, y):
    H = compute_H(x, y)
    #TODO Implement umeyama method and return R
    U, S, VT = np.linalg.svd(H)
    return np.asarray(VT.T * U.T)

# let's see the result
R_svd = spherical_umeyama(P_mov_demean, P_fix_demean)
R = R_svd
print("R =", R)

# Apply transformation (R and t) to original, moving dataset
t = mean_fix - np.dot(R, mean_mov) #TODO Compute translation
P_corr = np.dot(R, P_mov) + t

# Plot original fixed and corrected, original moving datasets -> should overlap
fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")
ax.scatter(P_fix[0,], P_fix[1,], P_fix[2,], c="r")
ax.scatter(P_corr[0,], P_corr[1,], P_corr[2,], c="b")
lim = 2
ax.set_xlim(-lim, lim)
ax.set_ylim(-lim, lim)
ax.set_zlim(-lim, lim)
plt.show()
```

## Part 2 - ICP

Until now, we have always assumed we know correspondence. Unfortunately, that is often not the case. The ICP algorithm is one of the most common ways to tackle this kind of registration problem.

For this part, we will use the face dataset above and just "forget" that we have point correspondance, simulating a realistic scanario. In order to do so, we just permute the points of the moving set.
Also, we will directly start with the demeaned datasets here - you should already know that's always the first step.

```python
# Data Permutation - we "forget" point correspondance
np.random.shuffle(P_mov_demean.T)
```

Okay, now the ICP algorithm itself:

```python
def icp(dst, src_init, iterations):
    # copy initial source points
    src = np.array(src_init, copy=True)
    
    R_total = np.identity(3) #TODO Initialize R. Start with identity transformation

    #TODO initialize nearest neighbor search tree with fixed points. 
    nbrs = NearestNeighbors(n_neighbors=1, algorithm="auto").fit(dst.T)
           #(Hint: use NearestNeighbors from sklearn.neighbors package)
    
    for i in range(iterations):
        #TODO select nearest neighbors to current point set
        distances, indices = nbrs.kneighbors(src.T)
        #TODO solve spherical registration between found correspondances (Umeyama)
        R = spherical_umeyama(src, dst[:, indices.T[0]])
        #TODO apply registration
        src = np.dot(R, src)
        
        # saving R to obtain a final transformation in the end
        R_total = np.dot(R, R_total);
        
    return (R_total, src)

######################################################
# Call ICP function
R_icp, P_mov_icp = icp(P_fix_demean, P_mov_demean, 20);

# Check whether we get the same result as before
print("R_sib =", R_svd, "\nR_icp =", R_icp)

# Plot result
fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")
ax.scatter(P_fix_demean[0,], P_fix_demean[1,], P_fix_demean[2,], c="r")
ax.scatter(P_mov_icp[0,], P_mov_icp[1,], P_mov_icp[2,], c="b")
lim = 2
ax.set_xlim(-lim, lim)
ax.set_ylim(-lim, lim)
ax.set_zlim(-lim, lim)
plt.show()
```

## Part 3 - Coherent Point Drift

*taken form https://siavashk.github.io/2017/05/14/coherent-point-drift/ and https://github.com/siavashk/pycpd/,
originally presented in Myronenko, Andriy, Xubo Song, and Miguel A. Carreira-Perpinn. "Non-rigid point set registration: Coherent point drift." Advances in neural information processing systems. 2007*

In CPD, the point cloud registration problem is solved in a probabilistic manner. It creates a distribution from the moving point cloud and assumes the fixed one is an observation from this distribution. This distribution is assumed to be a Gaussian Mixture Model. It tries to maximize the posterior probability of the observation by updating the distribution parameters, namely the mean and variance, which results in inferring the transformation.

- Gaussian Mixture Models example: (from https://scikit-learn.org/)

```python
from matplotlib.colors import LogNorm
from sklearn import mixture
%matplotlib inline

def GMM_example(n_samples=300):

    # generate random sample, two components
    np.random.seed(0)

    # generate spherical data centered on (20, 20)
    shifted_gaussian = np.random.randn(n_samples, 2) + np.array([20, 20])

    # generate zero centered stretched Gaussian data
    C = np.array([[0.0, -0.7], [3.5, 0.7]])
    stretched_gaussian = np.dot(np.random.randn(n_samples, 2), C)

    # concatenate the two datasets into the final training set
    X_train = np.vstack([shifted_gaussian, stretched_gaussian])

    # fit a Gaussian Mixture Model with two components
    clf = mixture.GaussianMixture(n_components=2, covariance_type="full")
    clf.fit(X_train)

    # display predicted scores by the model as a contour plot
    x = np.linspace(-20.0, 30.0)
    y = np.linspace(-20.0, 40.0)
    X, Y = np.meshgrid(x, y)
    XX = np.array([X.ravel(), Y.ravel()]).T
    Z = -clf.score_samples(XX)
    Z = Z.reshape(X.shape)

    CS = plt.contour(X, Y, Z, norm=LogNorm(vmin=1.0, vmax=1000.0), levels=np.logspace(0, 3, 10))
    CB = plt.colorbar(CS, shrink=0.8, extend="both")
    plt.scatter(X_train[:, 0], X_train[:, 1], 0.8)

    plt.title("Negative log-likelihood predicted by a GMM")
    plt.axis("tight")
    plt.show()


GMM_example()
```

## GMM based Registration
Two steps need to be done:
- find a correspondance based on current position of moving points
- find a transformation by maxtimizing the negative log-likelihood of the posterior probablity

**Solution: Expectation-Maiximization**

1. **E-step**:
find which Gaussian the observed point cloud is sampled from

<!--     $p(x_f) = \sum_{m=0}^{M} P(x_m)p(x_f|x_m)$ -->


2. **M-step**:
find the transformation using the infered correspondance from E-step

<!--     $E(\theta, \sigma^2) = -\sim_{j}^{N} \log $ -->

3. Update the points and go back to E-step until convergence


```python
import pycpd
from functools import partial
from mpl_toolkits.mplot3d import Axes3D
from pycpd import AffineRegistration


fish_target = np.loadtxt("data/fish_target.txt")
X1 = np.zeros((fish_target.shape[0], fish_target.shape[1] + 1))
X1[:, :-1] = fish_target
X2 = np.ones((fish_target.shape[0], fish_target.shape[1] + 1))
X2[:, :-1] = fish_target
X = np.vstack((X1, X2))

fish_source = np.loadtxt("data/fish_source.txt")
Y1 = np.zeros((fish_source.shape[0], fish_source.shape[1] + 1))
Y1[:, :-1] = fish_source
Y2 = np.ones((fish_source.shape[0], fish_source.shape[1] + 1))
Y2[:, :-1] = fish_source
Y = np.vstack((Y1, Y2))

fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c="r")
ax.scatter(Y[:, 0], Y[:, 1], Y[:, 2], c="b")
# lim=3; ax.set_xlim(2-lim, 2+lim); ax.set_ylim(-lim, lim); ax.set_zlim(-lim, lim)
plt.show()
```

```python
%matplotlib notebook

def visualize(iteration, error, X, Y, ax, fig):    
    ax.clear()
    ax.scatter(X[:, 0], X[:, 1], X[:, 2], color="red", label="Target")
    ax.scatter(Y[:, 0], Y[:, 1], Y[:, 2], color="blue", label="Source")
    ax.text2D(0.87, 0.92, "Iteration: {:d}\nQ: {:06.4f}".format(iteration, error),
              horizontalalignment="center", verticalalignment="center", transform=ax.transAxes, fontsize="x-large")
    ax.legend(loc="upper left", fontsize="x-large")
    fig.canvas.draw()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")

plt.ion()

fig.show()
fig.canvas.draw()

callback = partial(visualize, ax=ax, fig=fig)

reg = AffineRegistration(**{"X": X, "Y": Y})
reg.register(callback)
plt.show()
```

```python
from pycpd import DeformableRegistration


fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")

plt.ion()

fig.show()
fig.canvas.draw()

callback = partial(visualize, ax=ax, fig=fig)

reg = DeformableRegistration(**{"X": X, "Y": Y})
reg.register(callback)
plt.show()
```
