## Basic Segmentation (Thresholding)

We see the image and the histogram. Now we can start the implementation of a very basic segmentation method: thresholding. Do not forget to filter the image (with Gaussian filter) before segmenting. The segmentation will be done in the grayscale image.

```python
#Basic Segmentation (Thresholding)

def plot_seg(img, segmentation):
   # And now we show our results. 
    fig = plt.figure(figsize=(16, 5))
    ax1 = fig.add_subplot(121) 
    ax1.imshow(img, interpolation="none",  cmap="gray") 
    ax1.set_title("Original gray image") 

    ax2 = fig.add_subplot(122)
    ax2.imshow(segmentation, interpolation="none",  cmap="gray", vmin=0, vmax=255)
    ax2.set_title("Segmentation") 
    plt.show()


# A grayscale image img and a threshold value th. 
# return the segmented image.
def segment_th(img, ths):
    #TODO: Apply first Gaussian Blur to the image
    img = cv2.GaussianBlur(img, ksize=(25, 25), sigmaX=1)
    
    #TODO: let's create a black image (same size as our image_gray) to store our results. 
    # Segmentation by thresholding (can be done combining the operator "<" and the array element access "[]")
    segmented = np.zeros(shape=img.shape)
    segmented[img < ths] = 255

    #Plot the segentation
    plot_seg(img, segmented)


# TODO: Call your method here. Use the grayscale image. 
segment_th(image_gray, 150)
```

## K-Means

Now, let's work on an implementation of K-means. For this, we will again use the image of Trypanosoma cruzi parasite.

We need to complete the code for:
* A metric of distance or similarity
* A function for updating the means. 
* A function for updating the clusters. 
* A function that decides if we should stop.

Remember that the centers are defined in the color space. 

```python
# Here we select the locations of the image that will be used as initial centers (means)
def initial_centers(rows, cols, k):
    centerx = np.random.uniform(0, cols, k).astype(np.int32) 
    centery = np.random.uniform(0, rows, k).astype(np.int32)
    print(centerx)
    print(centery)
    return centerx, centery

# Define a similarity metric between two points (e.g l2). 
# x.shape=(3,), y.shape=(3,)
# Should return a single float
def similarity(x, y):
    #TODO: Compute the l2 distance between x and y
    dist = np.linalg.norm(x - y)
    return dist # your result

# Here we update the means. 
# We use the rgb (rows x cols x 3) image img
# A rows x cols matrix with the cluster map (clusters(i,j) in {0, 1, ..., k-1})
# And k the total of clusters (0, 1, ..., k-1)
# We return a k x 3 float32 array with the new means (centers) in each colum
def update_means(img, clusters, k):
    means = np.zeros(shape=(k, 3), dtype=np.float32)
    # TODO: Update the means according to the current clusters.
    for i in range(0, k):
        mean = np.mean(img[np.where(clusters == i)], axis=(0))
        means[i] = mean
    
    return means


# Here we update the clusters. 
# We use the rgb (rows x cols x 3) image img
# A k x 3 float32 array with the new means (centers) in each colum
# And k the total of clusters (0, 1, ..., k-1)
# We return a rows x cols matrix with the cluster map (clusters(i,j) in {0, 1, ..., k-1})
def update_clusters(img, means, k):
    rows, cols, channels = img.shape
    clusters = np.zeros(shape=(rows, cols))
    
    for r in range(rows): 
        for c in range(cols):
            min_dist = 90000
            p = img[r, c, :] # We get a 3d point shape=(3,)
            #TODO: use the similarity metric to find the cluster with the nearest mean to p
            # assign this cluster in the corresponding possition of the matrix "clusters"
            for i in range(0, k):
                current_dist = similarity(p, means[i])
                if current_dist < min_dist:
                    min_dist = current_dist
                    clusters[r, c] = i
            
    return clusters


# A simple stop criteria
def should_stop(meansi, meansim1, th=0.1):
    # TODO: compute the l1 distance (mean of abs values) between the mean at time i and means at time i - 1
    # Compare with th and return a boolean value indicating if we should stop. 
    stop = np.sum(np.abs(meansi - meansim1)) < th
    return stop # True or False
    

# The K. means alg. uses a rgb image and the number od clusters
def k_means(img, k):
    # Compute the localtions that will be used as initial centers
    cxs, cys = initial_centers(img.shape[0], img.shape[1], k)
    # Use the images intensities as the initial centers 
    means = np.zeros(shape=(k, 3), dtype=np.float32)
    for i in range(k):
        means[i] = img[cys[i], cxs[i], :]
    print(means)
    stop = False
    clusters = None
    while not stop:
        #TODO:
        # Backup your last means
        means_copy = np.copy(means)
        # Update the clusters
        clusters = update_clusters(img, means, k)
        # Update the means
        means = update_means(img, clusters, k)
        # Check the stop variable
        stop = should_stop(means, means_copy)
        
    print(means)  
    return clusters, means


# Just for visualization, we assign to each cluster the value of its mean (or center) 
def paint_clusters(shape, clusters, means, k):
    img = np.zeros(shape=shape, dtype=np.uint8)
    for i in range(k):
        img[np.where(clusters == i)] = np.round(means[i]).astype(np.uint8)
    return img


k = 3
denoised_image_color = cv2.GaussianBlur(image_color, ksize=(3, 3), sigmaX=3).astype(np.float32)
segmentation_km, means = k_means(denoised_image_color, k)
segmentation_km = paint_clusters(image_color.shape, segmentation_km, means, k)

fig = plt.figure(figsize=(16, 5))
ax1 = fig.add_subplot(121)
ax1.imshow(image_color) 
ax1.set_title("Input image")
ax2 = fig.add_subplot(122)
ax2.imshow(segmentation_km)
ax2.set_title("K Means") 

plt.show()
```

##  Comparing Pixel Distances

So far, the methods showed use only pixel-level information and no knowledge about the object (structure, animal, organ, etc) that we want to segment.

Now, let's use a more complex (but still basic) strategy that includes some information about the thing we want to segment: Color Information.

We can segment an image by comparing the distance of a pixel with a pre-computed model, e.g. a statistical model. This will lead to a distance map that we can then threshold. We can use different distance metrics, like the Euclidean distance or the Mahalanobis distance (https://en.wikipedia.org/wiki/Mahalanobis_distance). Let's see how this can be done:

```python
mask = cv2.imread("img/mask2.png", 1)
# A mask.
mask[mask > 0] = 1 

masked = image_color * mask # We use the mask to obtain samples for the object's pixels. 

fig = plt.figure(figsize=(16, 5)) 
ax1 = fig.add_subplot(121)
ax1.imshow(image_color)
ax1.set_title("Input image") 
ax2 = fig.add_subplot(122) 
ax2.imshow(masked) 
ax2.set_title("Masked Image")

plt.show()
```

```python
# We need the mean and variance of the object's pixels
image_mean = np.mean(masked[np.where(masked[:, :, 0] != 0)], axis=(0)).reshape(3, 1) 
image_variance = np.cov(masked[np.where(masked[:, :, 0] != 0)], rowvar=False)
image_variance = np.linalg.inv(image_variance) 
print(image_mean)
print(image_variance)
```

```python
def dist_segmentation(img, mean, inv_cov):
    rows, cols, chn = img.shape
    dist = np.zeros(shape=(rows, cols))
 
    for r in range(rows):
        for c in range(cols):
            p = img[r, c, :].reshape((3, 1))
            # TODO: compute the distance sqrt( (p-mean)t * inv_cov * (x-mean) ) and 
            # save it in the distance matrix (dist). * is the dot product. 
            diff = p - mean
            dist[r, c] = np.sqrt(np.dot(np.dot(diff.T, inv_cov), diff))
            
    return dist


denoised_image_color = cv2.GaussianBlur(image_color, ksize=(0, 0), sigmaX=2)
distances = dist_segmentation(denoised_image_color, image_mean, image_variance)
fig = plt.figure(figsize=(16, 5))
ax1 = fig.add_subplot(121) 
ax1.imshow(distances, interpolation="none",  cmap="gray") 
ax1.set_title("Distances image")
ax2 = fig.add_subplot(122) 
ax2.hist(distances.ravel(), bins=256)
ax2.set_title("Distances distribution") 

plt.show() 
```

## Region Growing

Finally, we will take a look at the region's growing methods. Let's think that our image (grayscale or color) is a graph, where the nodes are the pixels. Any particular pixel has a connection with each of its surrounding neighbors (connectivity 4 or 8). 

We can choose a particular pixel position to be the root (seed) of our graph (that now becomes a tree).  Then we can perform a breadth-first search on our tree (https://en.wikipedia.org/wiki/Breadth-first_search), but with an additional condition:

* To consider a node as a neighbor, it should have a small distance (in color space) with the root (seed).

Once we have discovered all the nodes that meet this condition, we will have our segmentation. This is the region growing algorithm. 

Let's complete the next code to do the segmentation: 

```python
# We first need a seed point (root), let's pick a random point based on our masks
non_zero = np.array(np.where(mask[:, :, 0] > 0))
point = int(np.random.uniform(0, non_zero.shape[1]))
seedy, seedx = non_zero[:, point]
print((seedy, seedx))
print(image_color[seedy, seedx, :])
```

```python
# Now, we need code to get the neighbor pixels
def get4ngb(rows, cols, x, y):
    # function ngb = get4ngb(rows, cols, x, y)
    # 
    # Input:
    # rows, cols    Number of rows and colums from the 2D Image
    # x,y           Coordinates of the center point
    #
    # Output:
    # ngb           list of possible neighbours

    # initialize empty result list
    ngb = list()

    # left
    if x > 1:
        ngb += [[y, x - 1]]

    # down
    if y < rows - 1:
        ngb += [[y + 1, x]]

    # right
    if x < cols - 1:
        ngb += [[y, x + 1]]

    # up
    if y > 1:
        ngb += [[y - 1, x]]

    return ngb 
```

Now, the algorithm itself. Here is where you need to do some code to complete the method (remember that the Y axis is related to the rows and the X to the cols of the image:

```python
# Our similarity metric. Given two points with shape=(3,), return true or false if the distance 
# is lower or biger than th. Use the l1 distance (mean of absolute values)
def is_similar(p1, p2, th):
    # TODO: your metric here.
    return np.mean(np.abs(p1 - p2)) < th


# The algorithm. Our root is the pixel in position (seedX; seedY)
def region_grow(I, seedX, seedY, th):
    if len(I.shape) == 3:
        rows, cols, ch = I.shape
    else:
        rows, cols = I.shape
    # our map for the nodes (pixels) that we have not discovered yet (-1),
    # we have discovered (0)
    # we have discovered and meet the similarity condition (1). This is our segmentation
    R = -np.ones(shape=(rows, cols))
    queue = []
    # enqueue the root
    queue.append([seedY, seedX])
    
    # Setting the root as discovered
    R[seedY, seedX] = 0
    
    #While our queue is not empty
    while len(queue) > 0:
        # we dequeue the first node of the queue
        v = queue[0]
        queue = queue[1:]
        # TODO: Use our similarity function to check the condition on the node v
        if is_similar(I[v[0], v[1]], I[seedY, seedX], th):
            # TODO: set the node v as "meet the condition", that is, R[vy, vx] = 1
            R[v[0], v[1]] = 1
            # TODO: determine neighbors
            nbgs = get4ngb(rows, cols, v[1], v[0])
            # for all the neighbors of v
            for i in range(0, len(nbgs)):
                qq = nbgs[i]
                # If it was not discovered yet:
                if R[qq[0], qq[1]] < 0:
                    #TODO: label as discovered
                    R[qq[0], qq[1]] = 0
                    #TODO: enqueue it 
                    queue.append([qq[0], qq[1]])
            
    return R
```

```python
def plot_region_grow(img, R):
    seed_img = np.copy(img)
    cv2.circle(seed_img, center=(seedx, seedy), radius=30, color=(255, 255, 255), thickness=2)
    fig = plt.figure(figsize=(16, 5)) 
    ax1 = fig.add_subplot(131) 
    ax1.imshow(cv2.GaussianBlur(seed_img, ksize=(0, 0), sigmaX=2), cmap="gray")
    ax1.set_title("Input image") 
    ax2 = fig.add_subplot(132) 
    ax2.imshow(R, interpolation="none",  cmap="gray") 
    ax2.set_title("Region Growed")


threshold = 10
denoised_image_color = cv2.GaussianBlur(image_color, ksize=(0, 0), sigmaX=2).astype(np.float32)
R = region_grow(denoised_image_color, seedx, seedy, threshold)
plot_region_grow(image_color, R)
```
